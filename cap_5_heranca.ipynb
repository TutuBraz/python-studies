```python
#Criando a super classe Animal
class Animal():
    
    def __init__(self):
        print('Animal criado')
        
    def identif(self):
        print('Animal')
        
    def comer(self):
        print('Comendo')
```

```python
#Criando a subclasse cachorro
class Cachorro(Animal):
    def __init__(self):
        Animal.__init__(self)
        print('O objeto cachorro foi criado')
        
    def identif(self):
        print('Cachorro')
        
    def latir(self):
        print('Au Au')
        
```

```python
rex = Cachorro()
```

```python
rex.comer()
```

```python
rex.latir()
```

```python
rex.identif()
```
