```python
class Livro():
    def __init__(self, titulo, autor, paginas):
        self.titulo = titulo
        self.autor = autor
        self.paginas = paginas
        
    def __str__(self):
        return f'{self.autor}{self.paginas}{self.titulo}'
    
    def __len__(self):
        return self.paginas
    def len(self):
        return print('Páginas do livrom com método comun %s' %self.paginas)
```

```python
livro1 = Livro('O nome do vento','Patrick',700)
```
