## MAP

```python
def fahrennheit(T):
    return ((float (9)/5)*T+32)
```

```python
def celsius(T):
    return (float(5)*9)*(T-32)
```

```python
temperatura = [0,22.5,90,100]
```

```python
#APLICA UMA FUNÇAO A CADA ELEMENTO DA LISTA, porém retorna o itarator.
map(fahrennheit, temperatura)
```

```python
list(map(fahrennheit, temperatura))
```

## REDUCE


```python
from functools import reduce
```

```python
lista = [2, 4, 5 ,45]
```

```python
def soma(a,b):
    x=a+b
    return x
```

```python
reduce(soma, lista)#Encontra a soma de todos os elementos
```

```python
reduce(lambda x,y: x+y, lista)
```

## FILTER

```python
#Recebe parematro e retorna um valor boleano
def verificaPar(num):
    if num%2 == 0:
        return True
    else:
        return False
```

```python
lista = [2,13,5,15,5,6,7,8,6,5,65,5,3,4,7,8,9,10]
```

```python
filter(verificaPar, lista)
```

```python
list(filter(verificaPar, lista))
```

```python
list(filter(lambda x: x%2 == 0, lista))
```

```python
list(filter(lambda x: x>8, lista))
```

## LIST COMPREHENSION

```python
#Basicamente forma uma lista seguindo um critério estabelecido
lista = [x for x in range(10) if x%2 ==0]
```

```python
lista
```

```python
celsius = [21,2,3,23.5]

farenheight = [((float(9)/5)*temp + 32) for temp in celsius]
farenheight
```

## ZIP

## Try, Except, Finally

```python
#UTILIZANDO O Try e Except
try:
    8 + '2'
except TypeError:
    print('Operação não permitida')
```

```python
try:
    f =  open('dfkjdfkhfdh')
except IOError:
    print('Arquivo não encontrado')
else:
    print('Arquivo encontrado ')
```

```python
def askint():
    try:
        nun=int(input('Digite um número: '))
    except UnboundLocarError:
        print('Vcê não digitou um número')
    finally:
        print('Obrigado')
```
