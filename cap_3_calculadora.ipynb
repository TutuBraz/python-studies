```python
print('Escolha uma operação a seguir: ')
print(' 1 - SOMA\n 2 - SUBTRAÇÃO\n 3 - MULTIPLICAÇÃO\n 4 - DIVISSSÃO')

option = int(input('Digite sua opção (1/2/3/4)'))
num1 = int(input('Digite o primeiro número: '))
num2 = int(input('Digite o segundo número: '))
if option == 1:
    print(f'{num1} + {num2} = {num1+num2}')
elif option == 2:
    print(num1-num2)
elif option == 3:
    print(num1*num2)
elif option == 4:
    print(num1/num2)
else:
    print('Não existe esse opção')
    
```
