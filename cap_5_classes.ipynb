```python
#criando classe
class Livro():
    #O método __init__ vai iniciar cada objeto a partir dessa classe.
    #self é uma referencia a cada atributo criado a partir dessa classe.
    def __init__(self): # método construtor
        self.titulo = 'O fim'
        self.isbn = 9988888
        print('Construtor chamado para criar um objeto desta classe')
        
    #métodos são funções, que recebem como parâmetro atributos doo objeto criado
    def imprime(self):
        print("Foi criado um livro %s ISBN %d" %(self.titulo, self.isbn))
        
```

```python
#Criando instancia da classe Livro
livro1 = Livro()
```

```python
#Tipo do objeto
type(livro1)
```

```python
#Atributo do objeto
livro1.isbn
```

```python
#Chamando método do objeto
livro1.imprime()
```

```python
#Criando classe Livro com parâmetros no método construtor
class Livro():
    def __init__(self, titulo, isbn):
        self.titulo = titulo
        self.isbn = isbn
        print('Construtor chamado para criar um objeto desta classe')
    
    def imprime(self, titulo, isbn):
        print(f'Foi criado um livro{titulo} ISBN {isbn}')
```

```python
livro2 = Livro('O nome do vento', 3943947)
```

```python
livro2.isbn
```

```python
livro2.imprime('A meninca')
```

```python
#Criando a classe Cachorro
class Cachorro():
    def __init__(self, raca):
        self.raca = raca
        print('Construtor chamado para criar um objeto desta classe')
```

```python
rex = Cachorro(raca='ViraLara')
```

```python
golias = Cachorro('Mel')
```

```python
rex.raca
```

```python
golias.raca
```
