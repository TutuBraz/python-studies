```python
#Criando a classe circulo
class Circulo():
    #Criando uma constante
    pi = 3.14
    
    #Quando o objeto dessa classe for criado, este metodo será executado o valor default do raio será 5
    def __init__(self, raio=5):
        self.raio = raio
        
    def area(self):
        return(self.raio*self.raio)*Circulo.pi
    
    def novoRaio(self, novo):
        self.raio = novo
        
    def getRaio(self):
        return self.raio
```

```python
circ = Circulo()
```

```python
circ.getRaio()
```

```python
circ.area()
```

```python
circ.novoRaio(8)
```

```python
circ.getRaio()
```
